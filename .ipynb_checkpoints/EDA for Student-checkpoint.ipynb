# Assigning Grades for better Analysis

```python
#Adding new column named Total
student_dataset['Total'] = student_dataset['math score'] + student_dataset['reading score'] +student_dataset['writing score']
student_dataset.head(1000)
```

```python
#Adding new column name Percentage
student_dataset['Percentage'] = (student_dataset['Total']/300) * 100
student_dataset.head(1000)
```

```python
#Assigning Grades(in CBSE pattern) according to percentage

def assign_grade(percentage):
    if percentage >= 81 and percentage <= 100:
        return 'Grade A'
    elif percentage >= 61 and percentage < 81:
        return 'Grade B'
    elif percentage >= 41 and percentage < 61:
        return 'Grade C'
    elif percentage >= 33 and percentage < 41:
        return 'Grade D'
    elif percentage >= 0 and percentage < 33:
        return 'Grade E'
    
student_dataset['Grades'] = student_dataset['Percentage'].apply(assign_grade)
student_dataset.head(1000)
```
